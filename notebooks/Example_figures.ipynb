```python
from ms_mint.io import ms_file_to_df
from glob import glob
```

```python
fn = '../static/demo_files/SA/SA_1.mzXML'
```

```python
df = ms_file_to_df(fn)
```

```python
figure(figsize=(8,4), dpi=300)

scatter(df.retentionTime, df['m/z array'], 
        c=df['intensity array'].apply(log1p), 
        s=0.1, marker='.', cmap='magma_r', vmax=10)

xlim(0,10)
ylim(100, 600)
grid()
colorbar(label='log intensity')
title('$\it{Staphylococcus\ aureus}$ sample')
plt.xlabel('Retention Time [min]')
plt.ylabel('M/Z')
savefig('../docs/image/demo_Saureus_sample_raw.png', dpi=300)
```

```python
figure(figsize=(8,4), dpi=300)

scatter(df.retentionTime, df['m/z array'], 
        c=df['intensity array'].apply(log1p), 
        s=df['intensity array'].apply(log1p), marker='.', cmap='magma_r', vmax=10)

xlim(.6,3)
ylim(117.019, 117.02)
grid()
colorbar(label='log intensity')
title('$\it{Staphylococcus\ aureus}$ sample (Succinate)')
plt.xlabel('Retention Time [min]')
plt.ylabel('M/Z')
gca().ticklabel_format(useOffset=False)
savefig('../docs/image/demo_Saureus_sample_raw_succinate.png', dpi=300)
```
