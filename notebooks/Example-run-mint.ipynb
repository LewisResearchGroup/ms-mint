## Check instrument mass-callibration with crosstab

For quality control purposes we can plot the values of the `peak_mass_diff_50pc` i.e. the median of the difference between observed data in the extraction window and the `mz_mean` valuein the target list. If the instrument is calibrated well these values should be around 0. The values are relative to the target mass in `ppm`. And since only data in the extraction window is considered the values are between `(-mz_width,mz_width )`. If the calibration is worth than this there will not be any intensity detected for the particular molecular target.

```python
mint.crosstab('peak_mass_diff_50pc').style.background_gradient(cmap='Reds')
```

```python
sns.catplot(data = mint.crosstab('peak_mass_diff_50pc').stack().to_frame().rename(columns={0: 'Mass calibration [ppm]'}).reset_index(), 
            x='ms_file_label', y='Mass calibration [ppm]', kind='violin', height=2.5, aspect=4)
```

We see that the _Candida_ samples (CA) have a different calibration than the rest of the files. The sets were probably generated at different points in time.

```python
(mint.crosstab('peak_rt_of_max') - mint.crosstab('rt')).T.style.background_gradient(cmap='Reds')
```

# Run Peak optimization

The peak optimization should be run using standards samples with a well defined concentration of the target metabolite to ensure the right peak is detected. Usually, expert knowledge is required in order to select the correct peak.

```python
mint.opt.rt_min_max(plot=True, sigma=200, rel_height=0.35)
```

The blue bar in this plot is the `rt` value given in the target list. The green dotted bar is the Gaussian weight that is applied to the data to select the correct peak. Sometimes, there is no one setting that works for all peaks simultaneously. Therefore, the peaks can be specified with the `peak_label` argument. For example: 

    mint.opt.rt_min_max(plot=True, sigma=200, rel_height=0.35, peak_label=['Arabitol'])
    
    

When the retention time range are optimized the data extraction process has to be repeated to update the results dataframe.
