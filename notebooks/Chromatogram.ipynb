```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import cwt, ricker

# Generate some synthetic data: a few peaks with different heights and some random baseline noise
x = np.linspace(0, 1000, 1000)
data = (1e6 * np.exp(-(x - 30)**2 / 50) +
        1e7 * np.exp(-(x - 500)**2 / 70) +
        1e5 * np.exp(-(x - 700)**2 / 50) +
        5e4 * np.random.rand(x.size) - 0.1 + 0.1)
```

```python
chrom = Chromatogram(scan_times=x, intensities=data, expected_rt=100)
```
